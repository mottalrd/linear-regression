```python
x = np.array([[-0.99768], [-0.69574], [-0.40373], [-0.10236], [0.22024], [0.47742], [0.82229]])
y = np.array([2.0885, 1.1646, 0.3287, 0.46013, 0.44808, 0.10013, -0.32952])
```

# Batch Gradient Regression

```python
reg = BatchGradient()
reg.fit(x, y)
```

```python
data = pd.DataFrame({'x':x.flatten(), 'y':y, 'pred': reg.predict(x)})
```

```python
%%R -i data -w 250 -h 250
ggplot(data=data, aes(x=x, y=y, pred=pred)) + 
    geom_point() +
    geom_point(aes(y=pred), color="green")
```
